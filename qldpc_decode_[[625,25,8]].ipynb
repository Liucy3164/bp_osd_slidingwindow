```python
import numpy as np
from bposd.hgp import hgp
from bposd.css_decode_sim import css_decode_sim
```

```python
h = np.loadtxt("/Users/louisliu/Desktop/New folder/research/Notes for QEC/5|-||-| QEC notes/bp_osd-main 3/examples/codes/classical_seed_codes/mkmn_20_5_8.txt").astype(int)
qcode = hgp(h)  # construct quantum LDPC code using the symmetric hypergraph product
qcode.test()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker

def sweep_lifetime_vs_p(
    qcode,
    Ws=(2, 3, 5, 10),
    F=1,
    ps=(0.01, 0.015, 0.02, 0.025, 0.03),
    target_runs=100,
    base_seed=42,
    xyz_error_bias=(np.inf, 0, 0),
    bp_method="minimum_sum",
    ms_scaling_factor=0.625,
    osd_method="osd_cs",
    osd_order=40,
    max_iter=1000,
    failure_decoder="osdw",
    meas_error_rate=None,
):
    """
    Runs len(Ws)*len(ps) sims and returns results dict: {W: [mean_lt(p1), ...]}.
    If meas_error_rate is None, it will default inside css_decode_sim to error_rate.
    """
    results = {int(W): [] for W in Ws}

    for W in Ws:
        W = int(W)
        for p in ps:
            p = float(p)
            osd_options = {
                "error_rate": p,
                "target_runs": int(target_runs),
                "xyz_error_bias": list(xyz_error_bias),
                "bp_method": bp_method,
                "ms_scaling_factor": float(ms_scaling_factor),
                "osd_method": osd_method,
                "osd_order": int(osd_order),
                "channel_update": None,
                # deterministic, different per point
                "seed": int(base_seed + 1000 * W + int(round(p * 10000))),
                "max_iter": int(max_iter),
                "failure_decoder": failure_decoder,
                "Window_W": W,
                "Offset_F": int(F),
            }
            if meas_error_rate is not None:
                osd_options["meas_error_rate"] = float(meas_error_rate)

            # NOTE: css_decode_sim must exist in your environment.
            sim = css_decode_sim(hx=qcode.hx, hz=qcode.hz, **osd_options)

            # css_decode_sim prints mean lifetime; we also read it programmatically:
            mean_lt = float(sim.mean_lifetime)
            results[W].append(mean_lt)

            print(f"(W,F)=({W},{int(F)}), p={p:.3f} -> mean lifetime {mean_lt:.6g}")

    return results


def plot_results(
    ps,
    results,
    Ws=(2, 3, 5, 10),
    title="Mean lifetime vs physical error rate",
):
    plt.figure()
    for W in Ws:
        W = int(W)
        ys = results[W]
        plt.plot(ps, ys, marker="o", linestyle="-")

    plt.xlabel("physical error rate (%)")
    plt.ylabel("mean lifetime")
    plt.title(title)

    plt.yscale("log", base=10)

    # ---- percentage x-axis ----
    plt.gca().xaxis.set_major_formatter(
        mticker.FuncFormatter(lambda x, _: f"{100*x:g}%")
    )

    plt.legend([f"(W,F)=({int(W)},1)" for W in Ws])
    plt.grid(True, which="both", linestyle="--", linewidth=0.5)
    plt.show()


# ---- Run points and plot ----
ps = [0.01, 0.015, 0.02, 0.025, 0.03]
Ws = [2, 3, 5, 10]

results = sweep_lifetime_vs_p(
    qcode=qcode,
    Ws=Ws,
    F=1,
    ps=ps,
    target_runs=100,
    base_seed=42,
    xyz_error_bias=(np.inf, 0, 0),
    bp_method="minimum_sum",
    ms_scaling_factor=0.625,
    osd_method="osd_cs",
    osd_order=40,
    max_iter=1000,
    failure_decoder="osdw",
    meas_error_rate=None,  # or set to 0.01 etc.
)

plot_results(
    ps,
    results,
    Ws=Ws,
    title="Mean lifetime vs error rate for (W,F) in {(2,1),(3,1),(5,1),(10,1)}",
)
```
